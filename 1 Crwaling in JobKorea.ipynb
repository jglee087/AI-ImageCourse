```python
import requests # from urllib.request import urlopen
from bs4 import BeautifulSoup

import pandas as pd
from datetime import datetime
import time
import re
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np 
```

```python
# 여러페이지는 그냥 url에 &Page_No= + nuber 하면 됨.
query = "영상처리"

urls_list = []
company_name = []
number = 1
while number <= 12:
    url = "http://www.jobkorea.co.kr/Search/?stext=" + query + "&Page_No=" + str(number)
    web = requests.get(url).content
    source = BeautifulSoup(web, 'html.parser')
    for urls in source.find_all('a', {'class' : "name"}):
        if "/C/" not in urls["href"] and "기업" not in urls["title"] and urls["title"] not in company_name:
            urls_list.append("http://www.jobkorea.co.kr" + urls["href"])
            company_name.append(urls["title"])
    number += 1

query = "영상SW"


number = 1
while number <= 2:
    url = "http://www.jobkorea.co.kr/Search/?stext=" + query + "&Page_No=" + str(number)
    web = requests.get(url).content
    source = BeautifulSoup(web, 'html.parser')
    for urls in source.find_all('a', {'class' : "name"}):
        if "/C/" not in urls["href"] and "기업" not in urls["title"] and urls["title"] not in company_name:
            urls_list.append("http://www.jobkorea.co.kr" + urls["href"])
            company_name.append(urls["title"])
    number += 1
    
query = "컴퓨터비전"


number = 1
while number <= 14:
    url = "http://www.jobkorea.co.kr/Search/?stext=" + query + "&Page_No=" + str(number)
    web = requests.get(url).content
    source = BeautifulSoup(web, 'html.parser')
    for urls in source.find_all('a', {'class' : "name"}):
        if "/C/" not in urls["href"] and "기업" not in urls["title"] and urls["title"] not in company_name:
            urls_list.append("http://www.jobkorea.co.kr" + urls["href"])
            company_name.append(urls["title"])
    number += 1  
print(len(urls_list))
```

```python
pass_char = ['/', '<', '>', 'd', 't', '\r', '\n']
for url in urls_list:
    print(url)
    web_news = requests.get(url).content
    source_news = BeautifulSoup(web_news, 'html.parser')
    
    
    cor_info = source_news.find_all('div', {'class' : 'coInfo'})
    cor_info = str(cor_info)
    cor_loc = source_news.find_all('div', {'class' : 'address'})
    cor_loc = str(cor_loc)
#     #이름
#     if "hd_4" in cor_info:
#         name_start = cor_info.find("hd_4") + 6
#     if "</h4>"in cor_info:
#         name_end = cor_info.find("</h4>") - 1
    
    #산업
    if "산업" in cor_info:
        temp = cor_info.find("산업") + 2
        while cor_info[temp] in pass_char:
            temp += 1
        industry_start= temp
        while cor_info[temp] not in pass_char:
            temp += 1
        industry_end = temp

        company_industry.append(cor_info[industry_start:industry_end].strip())
    else:
        company_industry.append("")
    print("산업 : " + cor_info[industry_start:industry_end].strip())
    
    #기업형태
    if "기업형태" in cor_info:
        temp = cor_info.find("기업형태") + 4
        while cor_info[temp] in pass_char:
            temp += 1
        form_start= temp
        while cor_info[temp] not in pass_char:
            temp += 1
        form_end = temp
        form = cor_info[form_start:form_end].strip()

    else:
        form = ""
    company_form.append(form)
    print("기업형태 : " + form)
    
    #위치
    if "mapDtl" in cor_loc:
        temp = cor_loc.find("mapDtl") + 8
        loc_start = temp
        while " " != cor_loc[temp]:
            temp += 1
        loc_end = temp
        loc = cor_loc[loc_start:loc_end]
    else:
        loc = ""
    company_location.append(loc)
    print("위치 : " + loc)
    time.sleep(3)
```

```python
word_dic = {}

for word in company_df['Industry']:
    if word == "":
        continue
    if word not in word_dic:
        word_dic[word] = 1 # changed from "0" to "1"
    else:
        word_dic[word] += 1
        
sorted_industry_dic = sorted(word_dic.items(), key=lambda x:x[1], reverse=True)
sorted_industry_dic
```
