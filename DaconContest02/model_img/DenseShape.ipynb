```python
from keras.models import Model
from keras.layers import Dense, Input, Add, BatchNormalization, Dropout, Activation, Concatenate
from keras.utils import plot_model

drop=0.2
```

```python
inputs= Input(shape=(20,))

x = Dense(128)(inputs)
x = Activation('relu')(x)

x = Dense(256)(x)
x = Activation('relu')(x)
x = Dropout(0.25)(x)

for i in range(3):
    xp = Dense(128)(x)

    x = Dense(256)(x)
    x = Activation('relu')(x)
    x= Dropout(0.25)(x)

    xq= Dense(128)(x)

    x= Concatenate()([xp,xq])
    x = Activation('relu')(x)

x= Dense(128)(x)
x= Dense(128)(x)

outputs1= Dense(19,activation='softmax')(x)

models= Model(inputs, outputs1)
```

```python
plot_model(models, 'fig_dense.png',show_shapes=True)
```
