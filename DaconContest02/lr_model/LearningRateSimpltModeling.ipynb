```python
import math
import matplotlib.pyplot as plt
import numpy as np

epoch=5000


#T_max= [50, 100, 150, 200]
T_max=100

eta_max=5e-4
eta_min=2e-5

#dec = 1, 2, 3, 4, 5
dec = 5

decay= epoch/dec

res=[]

for i in range(epoch):
    
    x= eta_min + (eta_max - eta_min)/2. * (1.+ np.cos(math.pi * np.mod(i,T_max) / (T_max))) * np.exp(-i/decay)
    res.append(x)

plt.plot(res)
#plt.savefig('lr2.png')
#plt.xlim(-1,2500)
```
