# Import Library

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline  
```

# Read Data

```python
train = pd.read_csv('./data/train.csv', index_col=0)
test = pd.read_csv('./data/test.csv', index_col=0)
sample_submission = pd.read_csv('./data/sample_submission.csv', index_col=0)
```

# Seperate the Data

```python
yy=train[['type']]
x=train.drop(columns=['type','fiberID'])
pred=test.drop(columns=['fiberID'])

col_name=x.columns
```

```python
path='./data/column_name.txt'

with open(path, 'w') as f:
    for i in range(len(col_name)):
        f.write(col_name[i])
        f.write('\n')    
```

# Boxplot

# DataFrame to NumPy array

```python
x_tmp = np.array(x)
y_tmp = np.array(yy)
```

# Original Plot

```python
from sklearn.preprocessing import StandardScaler

plt.figure(figsize=(25,25))

scaler1 = StandardScaler()

print('-'*80)

for i in range(0,20):
   
    xa_col=(x_tmp[:,i])
    xa_col=xa_col.reshape(-1,1)
    xa=xa_col

    plt.subplot(5,4,i+1)
    plt.hist(xa,bins=50)
    plt.gca().set_title(col_name[i])
    
    print(': {}  &  : {}'.format(col_name[i],xa.mean()))

plt.show()
```

# Apply StandardScaler and Plot

```python
from sklearn.preprocessing import StandardScaler

plt.figure(figsize=(25,25))

scaler1 = StandardScaler()

print('-'*80)

for i in range(0,20):
   
    x_col=(x_tmp[:,i])
    x_col=x_col.reshape(-1,1)
    scaler1.fit(x_col)

    xx_temp = scaler1.transform(x_col)

    plt.subplot(5,4,i+1)
    plt.hist(xx_temp,bins=50)
    plt.gca().set_title(col_name[i])
                
plt.show()
```

# Apply RobustScaler and Plot

```python
from sklearn.preprocessing import RobustScaler

plt.figure(figsize=(25,25))

scaler2 =  RobustScaler()

print('-'*80)

for i in range(0,20):
   
    x2_col=(x_tmp[:,i])
    x2_col=x2_col.reshape(-1,1)
    
    scaler2.fit(x2_col)
    x2_col = scaler2.transform(x2_col)

    plt.subplot(5,4,i+1)
    plt.hist(x2_col,bins=5)
    plt.gca().set_title(col_name[i])
                
plt.show()
```

# MaxAbsScaler

```python
from sklearn.preprocessing import MaxAbsScaler

plt.figure(figsize=(25,25))

scaler2 =  MaxAbsScaler()

print('-'*80)

for i in range(0,20):
   
    x2_col=(x_tmp[:,i])
    x2_col=x2_col.reshape(-1,1)
    
    scaler2.fit(x2_col)
    x2_col = scaler2.transform(x2_col)

    plt.subplot(5,4,i+1)
    plt.hist(x2_col,bins=5)
    plt.gca().set_title(col_name[i])
                
plt.show()
```
