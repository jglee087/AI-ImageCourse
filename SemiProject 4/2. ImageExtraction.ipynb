```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import os
import glob
```

```python
# 캐스케이드 파일 지정해서 검출기 생성하기
cascade_file = "haarcascade_frontface.xml"
cascade = cv2.CascadeClassifier(cascade_file)
```

```python

path='C:/Users/student/temp/ch'
```

```python
files1 = glob.glob(path + "\\/*.jpg")
files2 = glob.glob(path + "\\/*.png")
files3 = glob.glob(path + "\\/*.jpeg")
files4 = glob.glob(path + "\\/*.bmp")

files=files1+files2+files3+files4
```

```python
for i,f in enumerate(files):
    
    try:
        img=cv2.imread(f)
        img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        face_list = cascade.detectMultiScale(img_gray)
        
        for (x,y,w,h) in face_list:
            # print("얼굴의 좌표 =", x, y, w, h)
            # color = (0, 0, 255)
            # cv2.rectangle(img, (x, y), (x+w, y+h), color, thickness=10)
            
            roi = img[y:y+h, x:x+w]
            #plt.imshow(roi)
            #plt.show()
        
        cv2.imwrite(path+'/face/ch'+str(i)+".png", roi)
        #print('saved')   
        
    except:
        pass
```
