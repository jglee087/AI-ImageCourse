```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
from datetime import datetime
import time
import re
import pprint
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
search = '영상처리'
url_list = []
num = 1
url = f'http://www.saramin.co.kr/zf_user/search/recruit?searchword={search}&go=&flag=n&searchMode=1&searchType=&search_done=y&search_optional_item=n&recruitPage={num}&recruitSort=relation&recruitPageCount=40&inner_com_type=&company_cd=0,1,2,3,4,5,6,7,9'

web = requests.get(url).content
source = BeautifulSoup(web, 'html.parser')
link = source.find_all('div',{'class':'pagination'})
link = link[0].find_all('a')

#print(link)
```

```python
lastpages = []

for i in link :
    lastpages.append(i.attrs['page'])

last_page = max(lastpages)
#print(last_page)
```

```python
for i in range(1,int(last_page)+1) :
    url = f'http://www.saramin.co.kr/zf_user/search/recruit?searchword={search}&go=&flag=n&searchMode=1&searchType=&search_done=y&search_optional_item=n&recruitPage={i}&recruitSort=relation&recruitPageCount=40&inner_com_type=&company_cd=0,1,2,3,4,5,6,7,9'
    #print(url)
    web = requests.get(url).content
    source = BeautifulSoup(web, 'html.parser')
    company_info = source.find_all('h2',{'class':'job_tit'})
    
    for j in range(len(company_info)):
        link = company_info[j].find_all('a',{'target':'_blank'})
        first_article = link[0]
        str = 'http://www.saramin.co.kr'+first_article.attrs['href']
        url_list.append(str)
```

```python
name_list = []
sector_list = []
form_list = []
address_list = []

print(len(url_list))
```

```python


for i in range(len(url_list)-40):
    time.sleep(2)
    try:
        #path='C:/Users\jungg/OneDrive - 경희대학교/[AI 이미지 과정]/[강의자료]/4. Semi Project/SemiProject 1/TeamProject/3. Final/'
        driver = webdriver.Chrome('(driver) chromedriver.exe')
        driver.get(url_list[i]) # Get the page
        time.sleep(3)
        translated_contents = driver.find_element_by_class_name('wrap_info')

        pq=(translated_contents.text)
        ppq=pq.split('\n')

        temp1=ppq[0]#.split(' ')[0]
        temp2=ppq[2]
        temp3=ppq[6]
        temp4=ppq[-1]

        name_list.append(temp1)
        form_list.append(temp2)
        sector_list.append(temp3)
        address_list.append(temp4)
        
        print(i,end=' ')
        
        time.sleep(2)
        #print(i,temp1,temp2,temp3,temp4)
        driver.close()
        driver.quit()
    except:
        driver.close()
        driver.quit()
        #pass
    

    
df=pd.DataFrame({'name':name_list,'sector':sector_list,'form':form_list,'address':address_list})
```
